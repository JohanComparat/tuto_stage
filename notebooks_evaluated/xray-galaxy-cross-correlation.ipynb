## Overview

In this notebook we study the **angular cross-correlation between galaxies and soft X-ray photons** measured by the eROSITA all-sky survey ([Predehl et al. 2021](https://doi.org/10.1051/0004-6361/202039313); [Sunyaev et al. 2021](https://doi.org/10.1051/0004-6361/202141179)).  
When you cross-correlate the positions of galaxies with the distribution of X-ray photons over the sky, you measure how much X-ray emission is associated with those galaxies and their host dark matter halos.

The signal comes from **three physical components**:

| Component | Physical origin | Scale |
|-----------|----------------|-------|
| **Point sources** | AGN and X-ray binaries inside galaxies | $r < 80$ kpc |
| **Hot circumgalactic medium (CGM)** | Hot gas in and around dark matter halos | $20 < r < 2000$ kpc |
| **Two-halo term** | Large-scale structure clustering | $r > 2$ Mpc |

We will:
1. Load and explore the galaxy sample (LS10 BGS, $\log_{10} M_*/M_\odot > 10$)
2. Load the cross-correlation measurement from Comparat et al. 2025 (Zenodo)
3. Understand the halo occupation distribution (HOD) model
4. Model the angular cross-correlation with its decomposition into components
5. Compare the model to the data and to HOD theoretical predictions

## §1 Physical background

### 1.1 The eROSITA telescope and soft X-rays

**eROSITA** (extended ROentgen Survey with an Imaging Telescope Array) is a German–Russian X-ray telescope aboard the SRG spacecraft, launched in 2019.  
It operates in the **0.2–8 keV** energy range, with maximum sensitivity in the **0.5–2 keV soft band** ([Merloni et al. 2024](https://doi.org/10.1051/0004-6361/202346165)).  
The four all-sky scans (eRASS:4) used in Comparat et al. 2025 cover roughly **13 000 deg²** of the extragalactic sky accessible to the German consortium (eROSITA-DE).

In the soft X-ray band, photons are produced mainly by:
- **Thermal bremsstrahlung** (free-free emission) from hot plasma ($T \sim 10^6$–$10^8$ K)
- **Line emission** from highly ionized metals in hot gas
- **X-ray binaries (XRB)** — compact stellar remnants accreting from companion stars
- **Active galactic nuclei (AGN)** — supermassive black holes actively accreting

### 1.2 The galaxy sample: LS10 BGS

We use galaxies from the **Legacy Survey Data Release 10** (LS10; [Dey et al. 2019](https://doi.org/10.3847/1538-3881/ab089d)), selected following the DESI Bright Galaxy Sample (BGS) algorithm ([Hahn et al. 2023](https://doi.org/10.3847/1538-3881/acb212)):

- Magnitude cut: $r_{AB} < 19.5$
- Photometric redshifts from [Zhou et al. (2021)](https://doi.org/10.3847/1538-3881/ac0bb4)
- Stellar masses estimated with LePHare ([Ilbert et al. 2006](https://doi.org/10.1051/0004-6361:200463)) on $g,r,i,z,W1,W2$ bands
- Sample studied here: $\log_{10}(M_*/M_\odot) > 10$, $0.05 < z < 0.18$
- Total galaxies: **N = 2 759 238** over ~16 800 deg²

**Volume-limited** sample design means completeness above the stellar mass threshold is ~98% (Table 1 of Comparat et al. 2025).

### 1.3 The angular cross-correlation estimator

The galaxy–X-ray event angular cross-correlation $w^{\rm X-corr}_{\rm LS93}(\theta)$ is measured with the [Landy & Szalay (1993)](https://doi.org/10.1086/174515) estimator:

$$w^{\rm X-corr}_{\rm LS93}(\theta) = \frac{G \times E - G \times R - E \times R + R \times R}{R \times R}$$

where $G$ = galaxy positions, $E$ = X-ray event positions, $R$ = random catalog.  
The measurement is then converted to physical units [erg s$^{-1}$ kpc$^{-2}$] using the background surface brightness $S_X^R$ (Table 2 of the paper):

$$w^{\rm stack}_{\rm DP83}(R) = \frac{S_X^G - S_X^R}{S_X^R}$$

Both estimators are in excellent agreement for projected separations $20 < R < 300$ kpc (Fig. 4 of Comparat+2025).

### 1.4 The halo model framework

The **halo model** ([Cooray & Sheth 2002](https://doi.org/10.1016/S0370-1573(02)00276-4); [Asgari et al. 2023](https://doi.org/10.21105/joss.05648)) decomposes any clustering signal into:

- **1-halo term**: pairs of galaxies (or galaxy–X-ray pairs) within the **same** dark matter halo → dominates at small scales
- **2-halo term**: pairs in **different** halos → dominates at large scales, sensitive to large-scale bias

The **Halo Occupation Distribution (HOD)** tells us how many galaxies of a given type live in a halo of mass $M$. Following Comparat+2025 (after [Zheng et al. 2005](https://doi.org/10.1086/466510); [2007](https://doi.org/10.1086/518402)):

$$\langle N_c(M) \rangle = \frac{1}{2} \left[1 + \mathrm{erf}\left(\frac{\log_{10}M - \log_{10}M_{\rm min}}{\sigma_{\log M}}\right)\right]$$

$$\langle N_s(M) \rangle = \langle N_c(M) \rangle \left(\frac{M - M_0}{M_1'}\right)^{\alpha_{\rm sat}}$$

with tied parameters $\log_{10}(M_0) = \log_{10}(M_{\rm min}) + \Delta_{M_0}$ and $\log_{10}(M_1') = \log_{10}(M_0) + 1$.  
The best-fit HOD for the $\log_{10}(M_*/M_\odot) > 10$ sample (Table 3 of Comparat+2025):

| Parameter | Value |
|-----------|-------|
| $\log_{10} M_{\rm min}$ | $12.113^{+0.409}_{-0.257}$ |
| $\alpha_{\rm sat}$ | $1.184^{+0.059}_{-0.029}$ |
| $\sigma_{\log M}$ | $0.666^{+0.335}_{-0.236}$ |
| $\Delta_{M_0}$ | $0.052^{+0.182}_{-0.067}$ |

## §4 Loading the cross-correlation measurements

### 4.1 Angular cross-correlation $w(\theta)$

The Zenodo dataset provides the measured angular cross-correlation $w^{\rm X-corr}_{\rm LS93}(\theta)$ between galaxy positions and eROSITA soft X-ray events (0.5–2 keV, eRASS:4). The file contains:

- `theta` — angular separation [radians]
- `wtheta` — cross-correlation amplitude (dimensionless)
- `wtheta_err` — uncertainty from Jackknife resampling
- `convert_theta_to_kpc` — conversion factor [kpc/radian] at the mean redshift
- `beckground` — background surface brightness $S_X^R$ [erg s$^{-1}$ kpc$^{-2}$]

> **Note:** The column `beckground` is a typo (should be "background") inherited from the data reduction pipeline.

```python
# ── Load angular cross-correlation from Zenodo ───────────────────────────────
xcorr_file = (ZENODO_DIR / 'Measurements_Xcorr_Stacks' / 'XCORR' /
              f"LS10_VLIM_ANY_{SAMPLE['file_tag']}_GALxEVT_wtheta.fits")

with fits.open(xcorr_file) as f:
    xc = f[1].data
    theta_rad    = xc['theta'].copy()          # [rad]
    wtheta       = xc['wtheta'].copy()         # dimensionless
    wtheta_err   = xc['wtheta_err'].copy()     # 1-sigma
    conv_to_kpc  = xc['convert_theta_to_kpc'][0]   # kpc / rad at mean z
    bg_sx        = xc['beckground'][0]         # erg/s/kpc^2

theta_arcmin = np.degrees(theta_rad) * 60.0   # [arcmin]
R_kpc        = theta_rad * conv_to_kpc        # proper separation [kpc]

# Convert to physical surface brightness: S_X = w * S_X^R
Sx_data    = wtheta     * bg_sx               # [erg/s/kpc^2]
Sx_err     = wtheta_err * bg_sx

print(f"Background S_X^R = {bg_sx:.4e} erg/s/kpc^2")
print(f"Conversion at z={SAMPLE['zmean']}: 1 arcmin = {conv_to_kpc/60:.1f} kpc")
print(f"Angular range: {theta_arcmin.min():.3f} – {theta_arcmin.max():.1f} arcmin")
print(f"Physical range: {R_kpc.min():.0f} – {R_kpc.max():.0f} kpc")
```

The file provides surface brightness in annular bins (in physical kpc), separately for all galaxies (ANY), red sequence (RS) and blue cloud (BC).

```python
# ── Load stacked surface brightness profile from Zenodo ───────────────────────
stack_file = (ZENODO_DIR / 'Measurements_Xcorr_Stacks' / 'STACK' /
              f"LS10_VLIM_ANY_{SAMPLE['file_tag']}_BC_RS_frac.fits")

with fits.open(stack_file) as f:
    st = f[1].data
    R_min_kpc = st['R_min_kpc'].copy()
    R_mid_kpc = st['R_mid_kpc'].copy()
    R_max_kpc = st['R_max_kpc'].copy()
    Sx_mid    = st['Sx_ANY_mid'].copy()    # central estimate [erg/s/kpc^2]
    Sx_lo     = st['Sx_ANY_min'].copy()    # lower bound
    Sx_hi     = st['Sx_ANY_max'].copy()    # upper bound
    N_gal_st  = st['N_GAL'][0]             # number of galaxies in stack

Sx_err_stack = (Sx_hi - Sx_lo) / 2.0
print(f"N galaxies in stack: {N_gal_st:,}")
print(f"Radial bins: R_mid = {R_mid_kpc[:4]} ... kpc")
print(f"S_X range: {Sx_mid.min():.2e} – {Sx_mid.max():.2e} erg/s/kpc^2")
```

## §5 The HOD-extended X-ray cross-correlation model

### 5.1 Cosmology setup

Comparat et al. 2025 adopt the **Planck 2018 VI** cosmology ([Planck Collaboration VI 2020](https://doi.org/10.1051/0004-6361/201833910)):

| Parameter | Value |
|-----------|-------|
| $\Omega_c h^2$ | 0.11933 |
| $\Omega_b h^2$ | 0.02242 |
| $h$ | 0.6766 |
| $n_s$ | 0.9665 |
| $A_s$ | $2.105 \times 10^{-9}$ |
| $\sigma_8$ | $\approx 0.811$ |

The **linear matter power spectrum** is computed using CAMB ([Lewis et al. 2000](https://doi.org/10.1086/309179)).  
The **halo mass function** follows [Tinker et al. (2010)](https://doi.org/10.1088/0004-637X/724/2/878) with $\Delta = 200$ times the mean density.

```python
# ── Planck 2018 cosmology (Comparat+2025, Sect. 4.1.1) ──────────────────────
# Exact values from Table 1 of Planck Collaboration VI (2020)
theta_cosmo = dict(
    h         = 0.6736,
    Omega_b   = 0.0493,
    Omega_cdm = 0.2607,
    Omega_m   = 0.3100,
    n_s       = 0.9649,
    **{'ln10^{10}A_s': 3.044},
)
colossus_cosmo = dict(
    flat=True, H0=theta_cosmo['h']*100, Om0=theta_cosmo['Omega_m'],
    Ob0=theta_cosmo['Omega_b'], sigma8=0.811, ns=theta_cosmo['n_s'],
)
print("Planck 2018 cosmological parameters:")
for k, v in theta_cosmo.items():
    print(f"  {k:20s} = {v}")

# Derived quantities
h  = theta_cosmo['h']
Om = theta_cosmo['Omega_m']
cosmo_ap = FlatLambdaCDM(H0=h*100, Om0=Om)

# ── Power spectrum (CAMB-based, with interpolation cache for speed) ──────────
print("\nInitialising LinearPowerSpectrum (CAMB)...")
pk_lin = LinearPowerSpectrum()

class _CachedPkLin:
    """Thin interpolation cache around LinearPowerSpectrum to avoid repeated CAMB calls."""
    def __init__(self, obj, n_k=512):
        self._base   = obj
        self._k_ref  = np.logspace(-4, 1.5, n_k)
        self._log_k  = np.log(self._k_ref)
        self._cache  = {}
    def _key(self, z, th):
        return (round(float(z), 4), round(float(th['Omega_m']), 5),
                round(float(th['ln10^{10}A_s']), 4), round(float(th.get('h', 0.72)), 4))
    def pk_linear(self, k, z, theta):
        key = self._key(z, theta)
        if key not in self._cache:
            pk_ref = np.asarray(self._base.pk_linear(self._k_ref, float(z), theta))
            self._cache[key] = np.log(np.maximum(pk_ref, 1e-50))
        return jnp.asarray(np.exp(np.interp(
            np.log(np.asarray(k, dtype=float)), self._log_k, self._cache[key])))

pk_cached = _CachedPkLin(pk_lin)

# ── Halo mass function: Tinker et al. (2008) ────────────────────────────────
hmf = make_hmf('tinker08', pk_func=pk_cached.pk_linear)
halo_profile = HaloProfile(colossus_cosmo, cm_relation='duffy08')

# Pre-warm cache at the sample redshift
_ = pk_cached.pk_linear(np.array([0.1]), SAMPLE['zmean'], theta_cosmo)
print(f"HMF (Tinker+08) and power spectrum cache ready.")
```

### 5.2 HOD best-fit parameters

The HOD was fitted to the galaxy auto-correlation $w_p(r_p)$ (Fig. 2 of Comparat+2025). We use these fixed best-fit values when modeling the X-ray cross-correlation. The parametrization follows [Zheng et al. (2005)](https://doi.org/10.1086/510095).

$$\langle N_c(M) \rangle = \frac{1}{2}\left[1 + \mathrm{erf}\left(\frac{\log_{10}M - \log_{10}M_{\rm min}}{\sigma_{\log M}}\right)\right]$$

$$\langle N_s(M) \rangle = \langle N_c(M) \rangle \left(\frac{M - M_0}{M_1'}\right)^{\alpha_{\rm sat}} \mathcal{H}(M - M_0)$$

```python
# ── HOD best-fit (Table 3 of Comparat+2025, M*>10 sample) ──────────────────
HOD_BEST_FIT = dict(
    log10mmin  = 12.113,    # log10(M_min/[Msun/h])
    alpha_sat  = 1.184,     # satellite slope
    sigma_logm = 0.666,     # scatter in the central occupation
    delta_m0   = 0.052,     # log10(M0) = log10(Mmin) + delta_m0
)

# Derived parameters (Eq. 4–5 of Comparat+2025)
# log10(M0)  = log10(Mmin) + delta_m0
# log10(M1') = log10(M0) + 1  (standard 10:1 scaling)
log10m0 = HOD_BEST_FIT['log10mmin'] + HOD_BEST_FIT['delta_m0']
log10m1 = log10m0 + 1.0

# gga_model key names: 'log10m1' (not 'log10m1p') and 'alpha' (not 'alpha_sat')
hod_params = dict(
    log10mmin  = HOD_BEST_FIT['log10mmin'],
    sigma_logm = HOD_BEST_FIT['sigma_logm'],
    log10m0    = log10m0,
    log10m1    = log10m1,
    alpha      = HOD_BEST_FIT['alpha_sat'],
)

print("HOD parameters (from w_p fit, Table 3):")
for k, v in hod_params.items():
    print(f"  {k:15s} = {v:.3f}  ({10**v:.2e} Msun/h)" if 'log10m' in k else f"  {k:15s} = {v:.3f}")

# HODModel takes (hmf, hmf.bias) — the bias function is used for the 2-halo term
hod_model = HODModel(hmf, hmf.bias)

# ── Visualize the HOD ────────────────────────────────────────────────────────
log10m_arr = np.linspace(11, 15, 200)
nc_arr, ns_arr = hod_model.nc_ns(jnp.array(log10m_arr), hod_params)
nc_arr, ns_arr = np.array(nc_arr), np.array(ns_arr)

fig, ax = plt.subplots(figsize=(8, 5))
ax.semilogy(log10m_arr, nc_arr, 'b-', lw=2, label=r'$\langle N_c(M)\rangle$ — centrals')
ax.semilogy(log10m_arr, np.maximum(ns_arr, 1e-6), 'r--', lw=2,
            label=r'$\langle N_s(M)\rangle$ — satellites')
ax.semilogy(log10m_arr, np.maximum(nc_arr + ns_arr, 1e-6), 'k-', lw=2,
            label=r'$\langle N_{\rm tot}(M)\rangle$')
ax.axvline(HOD_BEST_FIT['log10mmin'], color='gray', ls=':', alpha=0.7,
           label=f"$M_{{\\rm min}} = 10^{{{HOD_BEST_FIT['log10mmin']:.2f}}}\\,M_\\odot/h$")
ax.set_xlabel(r'$\log_{10}(M_{200m}/[M_\odot\,h^{-1}])$', fontsize=12)
ax.set_ylabel(r'$\langle N(M) \rangle$', fontsize=12)
ax.set_title(r'HOD for $\log_{10}(M_*/M_\odot) > 10$ galaxies', fontsize=12)
ax.legend(fontsize=10)
ax.set_xlim(11.5, 15)
ax.set_ylim(1e-2, 20)
plt.tight_layout()
plt.show()
```

### 5.3 X-ray emissivity profile — the hot gas model

The X-ray emissivity profile of hot gas in a dark matter halo of mass $M_{200m}$ follows Eq. 8 of Comparat+2025 (adapted from [Vikhlinin et al. 2006](https://doi.org/10.1086/500143)):

$$\varepsilon(r) \propto \frac{1}{(r/r_s)^{\alpha_{\rm prof}} \left[1 + (r/r_s)^2\right]^{p_2}}$$

with fixed parameters $\alpha_{\rm prof} = 0.9$ and $p_2 = 1.6$ (consistent with cluster measurements from, e.g., [Neumann & Arnaud 1999](https://doi.org/10.1051/0004-6361:19990013); [Mantz et al. 2016](https://doi.org/10.1093/mnras/stw469)).  
The scale radius is $r_s = 0.3 \, r_{500c}$.

The corresponding **projected surface brightness** $S_X(R)$ is obtained via the Abel transform:

$$S_X(R) = 2 \int_0^{\sqrt{r_{\rm max}^2 - R^2}} \varepsilon\!\left(\sqrt{R^2 + l^2}\right) dl$$

This converts the 3D emissivity to the 2D brightness seen on the sky.

```python
# ── X-ray emissivity profile and its Abel projection ────────────────────────

# Show profiles for halos of different masses at z = 0.136
z_mean = SAMPLE['zmean']
halo_masses_msun_h = [1e13, 3e13, 1e14, 3e14]  # M_sun/h
halo_labels = [r'$10^{13}$', r'$3\times10^{13}$', r'$10^{14}$', r'$3\times10^{14}$']
colors_h = ['royalblue', 'darkorange', 'green', 'firebrick']

R_kpc_model = np.logspace(0.9, 3.3, 60)   # projected radii in kpc

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: 3D emissivity profile
ax = axes[0]
r_grid = np.logspace(-2, 1, 200)   # r / r_500c
eps = xray_gas_profile(r_grid, r_s=0.3, alpha_prof=0.9, p2=1.6)
ax.loglog(r_grid, eps / eps.max(), 'k-', lw=2)
ax.axvline(0.3, color='gray', ls=':', label='$r_s = 0.3\\,r_{500c}$')
ax.set_xlabel(r'$r / r_{500c}$', fontsize=12)
ax.set_ylabel(r'$\varepsilon(r)$ [normalized]', fontsize=12)
ax.set_title('3D X-ray emissivity profile (Eq. 8)', fontsize=12)
ax.set_xlim(0.01, 10)
ax.legend(fontsize=10)
ax.text(0.05, 0.5, r'$\alpha_{\rm prof}=0.9$, $p_2=1.6$', transform=ax.transAxes, fontsize=10)

# Right: Abel-projected surface brightness for different halo masses
ax = axes[1]
for m200m, label, col in zip(halo_masses_msun_h, halo_labels, colors_h):
    m500 = m500c_from_m200m_approx(m200m, z_mean, Om)
    r500_h = r500_from_m500(m500, z_mean, Om)       # Mpc/h
    r_s_h  = 0.3 * r500_h
    r_max_h = 5.0 * r500_h

    # Convert R_kpc to Mpc/h (comoving)
    R_mpc_h = (R_kpc_model / 1e3) * h * (1.0 + z_mean)
    sb = abel_project(R_mpc_h, r_s_h, r_max_h, n_los=128)

    # Scale by total expected luminosity
    d_l_mpc_h = float(cosmo_ap.luminosity_distance(z_mean).value) * h  # Mpc/h
    lx = lx_from_m500_comparat25(m500, z_mean, alpha_sr=1.63, om=Om)
    c_tot = counts_from_lx(lx, d_l_mpc_h)
    ax.loglog(R_kpc_model, sb * c_tot, '-', color=col, lw=2,
              label=f"$M_{{200m}} = {label}\\,M_\\odot/h$")

ax.set_xlabel(r'Projected radius $R$ [kpc]', fontsize=12)
ax.set_ylabel(r'$S_X(R)$ [counts (Mpc/h)$^{-2}$]', fontsize=12)
ax.set_title('Projected X-ray surface brightness per halo', fontsize=12)
ax.legend(fontsize=9)

plt.suptitle('Hot gas profile model (Comparat+2025, Eq. 8)', fontsize=13)
plt.tight_layout()
plt.show()
```

### 5.4 The $L_X$–$M_{500c}$ scaling relation

The X-ray luminosity of hot gas in a halo of mass $M_{500c}$ follows (Eq. 6 of Comparat+2025):

$$\log_{10} L_X^{0.5-2\,{\rm keV}} = L_0 + \alpha_{\rm SR} \left[\log_{10}\!\left(\frac{M_{500c}}{10^{15}\,M_\odot h^{-1}}\right) + \log_{10} E^2(z)\right]$$

where $E(z) = H(z)/H(0)$ and $L_0 = 44.7$ is fixed from cluster observations ([Seppi et al. 2022](https://doi.org/10.1051/0004-6361/202243824); [Bulbul et al. 2024](https://doi.org/10.1051/0004-6361/202347364)).  
The slope $\alpha_{\rm SR}$ is a **free parameter** fitted to the cross-correlation; the best fit gives $\alpha_{\rm SR} = 1.629^{+0.091}_{-0.089}$ (Table 4, $M_* > 10$ sample).

This steep slope ($\sim 1.63$) means that **massive halos contribute disproportionately** to the X-ray signal, even when they host fewer galaxies (see Fig. 13 of the paper).

```python
# ── L_X – M scaling relation ────────────────────────────────────────────────
m500_grid = np.logspace(12, 15, 100)   # M_sun/h
alpha_sr_best = 1.629                  # best-fit from cross-correlation (Table 4)

fig, ax = plt.subplots(figsize=(8, 5))

for alpha_sr, col, ls, label in [
    (1.629, 'k', '-',  r'Best-fit $\alpha_{\rm SR}=1.629$'),
    (1.4,   'b', '--', r'$\alpha_{\rm SR}=1.4$'),
    (1.9,   'r', '--', r'$\alpha_{\rm SR}=1.9$'),
]:
    lx_vals = lx_from_m500_comparat25(m500_grid, z=z_mean, alpha_sr=alpha_sr, om=Om)
    ax.loglog(m500_grid / 1e14, lx_vals / 1e44, color=col, ls=ls, lw=2, label=label)

ax.set_xlabel(r'$M_{500c}$ [$10^{14}\,M_\odot h^{-1}$]', fontsize=12)
ax.set_ylabel(r'$L_X^{0.5-2\,{\rm keV}}$ [$10^{44}$ erg s$^{-1}$]', fontsize=12)
ax.set_title(r'$L_X$–$M_{500c}$ scaling relation (Eq. 6 of Comparat+2025)', fontsize=12)
ax.legend(fontsize=10)
ax.set_xlim(0.01, 10)
ax.set_ylim(1e-4, 1e3)

plt.tight_layout()
plt.show()
```

### 5.5 Point source contribution

At small separations ($r < 80$ kpc), the cross-correlation is dominated by **X-ray point sources** — AGN and X-ray binaries residing inside galaxies. Their angular contribution is modeled as a PSF-convolved delta function at $\theta = 0$:

$$w^{\rm PSF}(\theta) = n_{\rm AGN} \times \mathrm{PSF}(\theta)$$

The eROSITA eRASS:4 PSF has FWHM $\approx 28$ arcsec ([Merloni et al. 2024](https://doi.org/10.1051/0004-6361/202346165)).  
The amplitude $\Delta_{M_{\rm min}}$ (a shift of the AGN HOD minimum mass) is a free parameter (Table 4: $\Delta_{M_{\rm min}} = 0.911^{+0.099}_{-0.101}$).

The **complete model** (Eq. 9 of the paper) is:

$$w^{\rm model}(\theta) = w^{\rm PSF}(\theta, \Delta_{M_{\rm min}}) + w^{\rm Gas \times Gal}(\theta, \alpha_{\rm SR})$$

where the gas term contains the 1-halo (CGM + satellite halos) and 2-halo (large-scale structure) contributions.

```python
# ── Illustrate the PSF (point source) component ──────────────────────────────
theta_fine_rad = np.logspace(np.log10(theta_rad.min()), np.log10(theta_rad.max()), 200)
theta_fine_arcmin = np.degrees(theta_fine_rad) * 60.0

# Point source amplitude: n_agn_per_gal = 0.011 (scaled from Delta_Mmin best-fit)
# (This is a representative value; actual fitting is done below)
n_agn_per_gal = 0.011
w_psf = w_psf_point_sources(theta_fine_rad, n_agn_per_gal, fwhm_arcsec=28.0)

fig, ax = plt.subplots(figsize=(8, 4))
ax.loglog(theta_fine_arcmin, w_psf, 'y-', lw=2.5, label='Point sources (AGN + XRB)')
ax.axvline(28.0/60.0, color='gray', ls=':', alpha=0.7, label='eROSITA PSF FWHM (28")')
ax.set_xlabel(r'Angular separation $\theta$ [arcmin]', fontsize=12)
ax.set_ylabel(r'$w^{\rm PSF}(\theta)$ contribution', fontsize=12)
ax.set_title('Point source contribution (PSF-smeared)', fontsize=12)
ax.legend(fontsize=10)
ax.set_xlim(0.06, 60)
plt.tight_layout()
plt.show()
```

## §6 Computing the HOD-weighted stacked profile

The **mean stacked X-ray surface brightness per galaxy** is the HOD-weighted sum over all halos:

$$\langle S_X \rangle(R) = \frac{\displaystyle\int dM \,\frac{dn}{dM}\, N_{\rm gal}(M)\, C_X(M)\, S_X^{\rm shape}(R \mid M)}{\displaystyle\int dM\, \frac{dn}{dM}\, N_{\rm gal}(M)}$$

where:
- $dn/dM$ is the halo mass function ([Tinker et al. 2010](https://doi.org/10.1088/0004-637X/724/2/878))
- $N_{\rm gal}(M)$ is the HOD (sum of centrals + satellites)
- $C_X(M)$ is the total expected X-ray photon count from a halo of mass $M$ (using Eq. 6–7)
- $S_X^{\rm shape}(R \mid M)$ is the normalized Abel-projected profile (integral over area = 1)

> **Key insight:** The integral over halo mass picks up contributions from both low-mass halos (many but faint) and high-mass halos (few but bright). The steep $L_X$–$M$ slope means the signal at large radii is dominated by rare massive halos hosting satellite galaxies (Fig. 13 of Comparat+2025).

This computation can take a few minutes depending on the number of mass grid points and line-of-sight steps.

```python
# ── Compute HOD-weighted stacked X-ray profile ───────────────────────────────
z_mean    = SAMPLE['zmean']
d_l_mpc_h = float(cosmo_ap.luminosity_distance(z_mean).value) * h  # Mpc/h
alpha_sr  = 1.629    # best-fit slope (Table 4)

R_model_kpc = np.logspace(0.9, 3.2, 30)  # kpc (30 points is sufficient for illustration)

print("Computing HOD-weighted surface brightness profile...")
print(f"  z = {z_mean}, d_L = {d_l_mpc_h:.1f} Mpc/h, alpha_SR = {alpha_sr}")

sx_model_gas = hod_weighted_sx_profile(
    R_kpc       = R_model_kpc,
    z           = z_mean,
    d_l_mpc_h   = d_l_mpc_h,
    hmf         = hmf,
    hod         = hod_model,
    hod_params  = hod_params,
    theta_cosmo = theta_cosmo,
    alpha_sr    = alpha_sr,
    r_s_fac     = 0.30,
    r_max_fac   = 5.0,
    log10m_min  = 11.5,
    log10m_max  = 15.5,
    n_mass      = 25,
    n_los       = 128,
)

# Convert from counts/(Mpc/h)^2 to erg/s/kpc^2 using background
# S_X [erg/s/kpc^2] = S_X [counts/(Mpc/h)^2] * bg_sx [erg/s/kpc^2]
# The stacked data is in physical erg/s/kpc^2; model output needs scaling by background
sx_model_erg = sx_model_gas * bg_sx

print(f"Profile computed: R = {R_model_kpc.min():.0f}–{R_model_kpc.max():.0f} kpc")
print(f"S_X model range: {sx_model_erg.min():.2e} – {sx_model_erg.max():.2e} erg/s/kpc^2")
```

## §7 Angular cross-correlation prediction

The angular cross-correlation function $w(\theta)$ is computed from the angular power spectrum $C_\ell^{gX}$ via the Limber approximation ([Limber 1953](https://doi.org/10.1086/145672)):

$$C_\ell^{gX} = \int dz \frac{c}{H(z)} \frac{1}{\chi^2(z)} \left[I^{gX}_{\rm 1h}(\ell, z) + I^{gX}_{\rm 2h}(\ell, z)\right]$$

where the 1-halo term is:
$$I^{gX}_{\rm 1h} = \frac{1}{\bar{n}_g} \int dM \frac{dn}{dM} N_{\rm gal}(M) \tilde{S}_X(\ell, M, z)$$

and $\tilde{S}_X$ is the 2D Fourier transform of the projected X-ray profile.  
The 2-halo term couples through the galaxy bias $b_g$ and the linear matter power spectrum $P_{\rm lin}(k)$.

Finally, $w(\theta)$ is recovered by a flat-sky Hankel transform ([Asgari et al. 2023](https://doi.org/10.21105/joss.05648)):
$$w(\theta) = \frac{1}{2\pi} \int_0^\infty \ell\, C_\ell^{gX}\, J_0(\ell\theta)\, d\ell$$

> **Computational note:** The full 1h+2h prediction requires integration over redshift, mass, and multipoles. We use a coarse grid here (N_mass=20, N_ell=25) for speed; finer grids improve accuracy at ~5% cost.

```python
# ── Angular cross-correlation prediction via Limber ──────────────────────────
ell_arr  = np.logspace(1, 4, 25)           # multipoles ℓ
z_grid   = np.linspace(SAMPLE['z_min'] + 0.005,
                        SAMPLE['z_max'] - 0.005, 6)   # redshift integration grid

print(f"Computing C_ell^{{gX}} on {len(ell_arr)} ell points across z = {z_grid}")

# Use the cached linear power spectrum to avoid repeated CAMB calls
def pk_lin_func(k, z, theta):
    return pk_cached.pk_linear(k, z, theta)

cl_gx = cl_gal_xray(
    ell_arr     = ell_arr,
    z_grid      = z_grid,
    hmf         = hmf,
    pk_lin_func = pk_lin_func,
    hod         = hod_model,
    hod_params  = hod_params,
    theta_cosmo = theta_cosmo,
    alpha_sr    = alpha_sr,
    log10m_min  = 11.5,
    log10m_max  = 15.5,
    n_mass      = 20,
    n_r         = 64,
)

# Hankel transform C_ell → w(theta)
w_gas = wtheta_from_cl(theta_rad, ell_arr, cl_gx)

# Point source contribution
w_ps = w_psf_point_sources(theta_rad, n_agn_per_gal=n_agn_per_gal, fwhm_arcsec=28.0)

# Total model
w_total = w_gas + w_ps

print("Prediction complete.")
```

## §10 Summary and discussion

### What we have done

1. **Loaded** a volume-limited galaxy sample (LS10 BGS, $\log_{10}(M_*/M_\odot) > 10$, $z < 0.18$, N = 2.76M).
2. **Loaded** the Comparat+2025 cross-correlation measurement between these galaxies and eROSITA soft X-rays (Zenodo: [10.5281/zenodo.15111974](https://doi.org/10.5281/zenodo.15111974)).
3. **Explained** the physical origin of each signal component:
   - Point sources (AGN + XRB) dominate at $R < 80$ kpc
   - Hot CGM emission dominates at $20 < R < 2000$ kpc for high-mass samples
   - Two-halo term dominates at $R > 2$ Mpc
4. **Modeled** the signal using the extended HOD framework of Comparat+2025 implemented in `gga_model`.
5. **Compared** the HOD auto-correlation prediction ($w_p$) with observations using `hod_mod`.

### Key physical insights

- The cross-correlation amplitude **increases with stellar mass threshold** — more massive galaxies live in more massive halos, which are brighter in X-rays.
- The **steep $L_X$–$M$ slope** ($\alpha_{\rm SR} \approx 1.63$) means a few massive halos dominate the large-scale signal, even when they host few galaxies.
- The **point source term** encodes the AGN population without the need for AGN-selected samples — this is a new way to study average AGN activity as a function of host galaxy mass.
- The tight constraint on $\alpha_{\rm SR}$ (over 3 orders of magnitude in halo mass!) is unprecedented at this level of precision.
