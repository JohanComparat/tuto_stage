## 1. Theoretical background

### 1.1 The Planck blackbody function

A **blackbody** (or black body) is an idealised object that absorbs all incident radiation and emits radiation with a spectrum determined solely by its temperature $T$. Stars are well approximated as blackbodies in many contexts.

Max Planck (1900) derived the spectral radiance (energy emitted per unit time, per unit area, per unit solid angle, per unit frequency interval):

$$B_\nu(T) = \frac{2h\nu^3}{c^2} \frac{1}{e^{h\nu/(k_B T)} - 1}$$

where:
- $h = 6.626 \times 10^{-34}$ J s — Planck's constant
- $c = 2.998 \times 10^8$ m s$^{-1}$ — speed of light in vacuum
- $k_B = 1.381 \times 10^{-23}$ J K$^{-1}$ — Boltzmann constant
- $\nu$ — photon frequency [Hz]
- $T$ — absolute temperature [K]

Units of $B_\nu$: W m$^{-2}$ Hz$^{-1}$ sr$^{-1}$.

The equivalent expression in **wavelength** is obtained using the relation $|B_\lambda \, d\lambda| = |B_\nu \, d\nu|$ and $\nu = c/\lambda$:

$$\boxed{B_\lambda(T) = \frac{2hc^2}{\lambda^5} \frac{1}{e^{hc/(\lambda k_B T)} - 1}}$$

Units of $B_\lambda$: W m$^{-2}$ m$^{-1}$ sr$^{-1}$ = W m$^{-3}$ sr$^{-1}$.

**Note:** The two functions $B_\nu$ and $B_\lambda$ peak at *different* wavelengths (frequencies) because of the Jacobian of the coordinate transformation.

### 1.2 Wien's displacement law

Differentiating $B_\lambda$ with respect to $\lambda$ and setting the result to zero gives the wavelength of peak emission:

$$\lambda_\mathrm{max} \cdot T = b_W \equiv 2.897771955 \times 10^{-3} \text{ m·K}$$

This is **Wien's displacement law** (NIST CODATA 2018 value). The peak shifts to shorter (bluer) wavelengths for hotter stars:
- M dwarf at $T = 3000$ K: $\lambda_\mathrm{max} \approx 966$ nm (near-infrared)
- Sun at $T = 5778$ K: $\lambda_\mathrm{max} \approx 501$ nm (green-yellow visible)
- O star at $T = 30000$ K: $\lambda_\mathrm{max} \approx 97$ nm (far-ultraviolet)

### 1.3 The Stefan–Boltzmann law

Integrating $B_\lambda$ over all wavelengths and solid angles gives the total flux emitted per unit surface area:

$$F = \sigma_{\rm SB} T^4 \quad \text{[W m}^{-2}\text{]}$$

where $\sigma_{\rm SB} = 5.670374419 \times 10^{-8}$ W m$^{-2}$ K$^{-4}$ (NIST CODATA 2018).

For a spherical star of radius $R$ and uniform surface temperature $T$:

$$\boxed{L = 4\pi R^2 \sigma_{\rm SB} T^4}$$

This is the **Stefan–Boltzmann law** for stellar luminosity.

### 1.4 The photometric magnitude system

Astronomy uses a **logarithmic scale** for brightness, with magnitudes defined such that a factor of 100 in flux corresponds to exactly 5 magnitudes:

$$m_1 - m_2 = -2.5 \log_{10}\!\left(\frac{F_1}{F_2}\right)$$

Photometric observations are always done through **filters** (bandpasses). The two classical filters relevant here are:
- **B band** (blue): centred near 440 nm
- **V band** (visual): centred near 550 nm

The **B−V colour index** is simply the difference of magnitudes in these two bands:

$$B-V = m_B - m_V = -2.5 \log_{10}\!\left(\frac{F_B}{F_V}\right) + C_{BV}$$

where $C_{BV}$ is a constant chosen so that $B-V = 0$ for a star of spectral type A0 (Vega). Hotter stars are bluer and have smaller (more negative) $B-V$; cooler stars are redder and have larger $B-V$.

### 1.5 The Ballesteros (2012) formula

Ballesteros (2012, EPL 97, 34008; arXiv:1201.1809) derived an analytic approximation relating the $B-V$ colour index to effective temperature, based on Planck's law integrated through idealised rectangular bandpasses:

$$\boxed{T_\mathrm{eff} = 4600 \left( \frac{1}{0.92(B-V) + 1.7} + \frac{1}{0.92(B-V) + 0.62} \right) \text{ K}}$$

This formula is remarkably accurate (to $\sim$5%) across the full range of stellar temperatures.

### 1.6 Stellar spectral types

Stars are classified into spectral types O–B–A–F–G–K–M in order of decreasing temperature (mnemonic: *Oh Be A Fine Girl/Guy, Kiss Me*). Key values from Pecaut & Mamajek (2013), ApJS 208, 9:

| Spectral type | $T_\mathrm{eff}$ [K] | Colour | Example |
|:---:|:---:|:---:|:---:|
| O5 | 42000 | blue-violet | Zeta Puppis |
| B0 | 30000 | blue-white | Spica |
| A0 | 9750 | white | Sirius A |
| F0 | 7300 | yellow-white | Procyon A |
| G0 | 5940 | yellow | Sun (G2) |
| K0 | 5270 | orange | Epsilon Eridani |
| M0 | 3850 | red-orange | Proxima Centauri |

## 3. The Planck function: $B_\lambda$ and $B_\nu$

We define the two forms of the Planck function and verify they are related by the Jacobian $B_\lambda = B_\nu \cdot c/\lambda^2$.

```python
def B_lambda(lam, T):
    """
    Planck spectral radiance per unit wavelength.

    Parameters
    ----------
    lam : array-like, float
        Wavelength [m]
    T : float
        Temperature [K]

    Returns
    -------
    B : array-like
        Spectral radiance [W m^-2 m^-1 sr^-1]
    """
    lam = np.asarray(lam, dtype=float)
    exponent = h * c / (lam * k_B * T)
    # Guard against overflow for very short wavelengths / low temperatures
    exponent = np.minimum(exponent, 700.0)
    return (2.0 * h * c**2 / lam**5) / (np.exp(exponent) - 1.0)


def B_nu(nu, T):
    """
    Planck spectral radiance per unit frequency.

    Parameters
    ----------
    nu : array-like, float
        Frequency [Hz]
    T : float
        Temperature [K]

    Returns
    -------
    B : array-like
        Spectral radiance [W m^-2 Hz^-1 sr^-1]
    """
    nu = np.asarray(nu, dtype=float)
    exponent = h * nu / (k_B * T)
    exponent = np.minimum(exponent, 700.0)
    return (2.0 * h * nu**3 / c**2) / (np.exp(exponent) - 1.0)


# ── Verify the Jacobian relation B_lambda = B_nu * (c / lambda^2) ──────────
lam_test = 500e-9   # 500 nm
nu_test  = c / lam_test

val_lam = B_lambda(lam_test, T_sun)
val_nu  = B_nu(nu_test, T_sun)
val_converted = val_nu * c / lam_test**2   # B_nu -> B_lambda

print('Verification of Jacobian B_lambda = B_nu * c / lambda^2')
print(f'  B_lambda(500 nm, T_sun)           = {val_lam:.4e} W m^-3 sr^-1')
print(f'  B_nu(nu_500nm, T_sun) * c/lam^2   = {val_converted:.4e} W m^-3 sr^-1')
print(f'  Relative difference: {abs(val_lam - val_converted)/val_lam * 100:.6f}%')
```

```python
# Wavelength grid: 100 nm to 3000 nm
lam_nm  = np.linspace(100, 3000, 5000)      # [nm]
lam_m   = lam_nm * 1e-9                     # [m]

# Temperatures to plot
temps   = [3000, 5778, 10000, 30000]
colors  = ['firebrick', 'goldenrod', 'steelblue', 'mediumpurple']
labels  = ['3 000 K (M dwarf)', '5 778 K (Sun, G2)', '10 000 K (A star)', '30 000 K (B star)']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax_idx, (ax, norm) in enumerate(zip(axes, [False, True])):

    # Shade the visible spectrum (380–700 nm)
    ax.axvspan(380, 700, color='lightyellow', alpha=0.8, zorder=0, label='Visible (380–700 nm)')

    for T, col, lbl in zip(temps, colors, labels):
        B = B_lambda(lam_m, T) * 1e-9   # convert to W m^-2 nm^-1 sr^-1

        if norm:
            B = B / B.max()   # normalise to peak = 1

        ax.plot(lam_nm, B, color=col, lw=2, label=lbl)

        # Mark Wien peak
        lam_peak_nm = b_Wien / T * 1e9
        if 100 < lam_peak_nm < 3000:
            B_peak = B_lambda(lam_peak_nm * 1e-9, T) * 1e-9
            if norm:
                B_peak = 1.0
            ax.axvline(lam_peak_nm, color=col, lw=0.8, ls='--', alpha=0.7)
            if ax_idx == 1:  # only annotate on right panel
                ax.annotate(f'{lam_peak_nm:.0f} nm',
                            xy=(lam_peak_nm, B_peak),
                            xytext=(lam_peak_nm + 80, B_peak - 0.05),
                            fontsize=8, color=col,
                            arrowprops=dict(arrowstyle='->', color=col, lw=0.8))

    ax.set_xlabel('Wavelength $\\lambda$ [nm]')
    if norm:
        ax.set_ylabel('Normalised spectral radiance')
        ax.set_title('Normalised Planck spectra (peak = 1)')
    else:
        ax.set_ylabel('$B_\\lambda$ [W m$^{-2}$ nm$^{-1}$ sr$^{-1}$]')
        ax.set_title('Planck blackbody spectra')
    ax.set_xlim(100, 2500)
    ax.legend(fontsize=9, loc='upper right')
    ax.xaxis.set_minor_locator(AutoMinorLocator())

plt.tight_layout()
plt.savefig('blackbody_spectra.png', dpi=150, bbox_inches='tight')
plt.show()
print('Left: absolute spectra — note 30 000 K peak is off the plot (in UV).')
print('Right: normalised spectra — peak wavelength shifts clearly with temperature.')
```

## 6. Stefan–Boltzmann law: total luminosity

The total luminosity of a star with radius $R$ and effective temperature $T$ is:

$$L = 4\pi R^2 \sigma_{\rm SB} T^4$$

To compute $L$ for each spectral type, we need both $T_\mathrm{eff}$ and the **main-sequence radius** $R$. We use tabulated values from Pecaut & Mamajek (2013).

```python
def stefan_boltzmann_luminosity(R_Rsun, T_K):
    """
    Compute stellar luminosity using the Stefan-Boltzmann law.

    Parameters
    ----------
    R_Rsun : float
        Stellar radius in solar units
    T_K : float
        Effective temperature in Kelvin

    Returns
    -------
    L_W : float
        Luminosity in Watts
    L_Lsun : float
        Luminosity in solar units
    """
    R_m    = R_Rsun * R_sun
    L_W    = 4.0 * np.pi * R_m**2 * sigma_SB * T_K**4
    L_Lsun = L_W / L_sun
    return L_W, L_Lsun


# Compute L for each spectral type and for a range of radii/temperatures
print('Luminosity for each spectral type (main-sequence radii from Pecaut & Mamajek 2013):')
print(f'\n{"Spt":>4}  {"T_eff [K]":>10}  {"R [R_sun]":>10}  {"L [W]":>12}  {"L/L_sun":>10}  {"lambda_max [nm]":>16}')
print('-' * 75)

luminosities = {}
for spt, (Teff, BV, R_R) in SPTYPES.items():
    L_W, L_ratio = stefan_boltzmann_luminosity(R_R, Teff)
    lam_p = b_Wien / Teff * 1e9
    luminosities[spt] = (Teff, R_R, L_W, L_ratio)
    print(f"{spt:>4}  {Teff:>10.0f}  {R_R:>10.2f}  {L_W:>12.3e}  {L_ratio:>10.3f}  {lam_p:>16.1f}")

print(f"\nSun:  T={T_sun:.0f} K, R=1.00 R_sun, L={L_sun:.3e} W, L/L_sun=1.000")
print(f"  (S-B formula gives: {stefan_boltzmann_luminosity(1.0, T_sun)[0]:.3e} W)")

# Show L ∝ T^4 scaling for fixed radius
print('\nIllustration: doubling T increases L by factor 2^4 = 16')
T_ref = 5000.0
L_ref, _ = stefan_boltzmann_luminosity(1.0, T_ref)
for fac in [0.5, 1.0, 2.0, 3.0]:
    L_fac, _ = stefan_boltzmann_luminosity(1.0, T_ref * fac)
    print(f'  T = {T_ref*fac:.0f} K  =>  L/L_ref = {L_fac/L_ref:.1f}')
```

```python
# Plot: L/Lsun vs T_eff for fixed radius (R = 1 R_sun) to illustrate T^4 law
# and for the spectral-type data points
T_arr = np.linspace(2500, 50000, 500)
L_fixed_R, _ = stefan_boltzmann_luminosity(np.ones_like(T_arr), T_arr)
L_fixed_R_norm = L_fixed_R / L_sun

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left panel: L/Lsun vs T for fixed R=1 Rsun
ax = axes[0]
ax.plot(T_arr, L_fixed_R_norm, 'darkblue', lw=2.5,
        label='$R = 1\\,R_\\odot$ (Stefan–Boltzmann $L \\propto T^4$)')
ax.scatter([T_sun], [1.0], color='gold', s=200, zorder=10,
           edgecolors='k', lw=0.8, label='Sun')
ax.set_xlabel('Effective temperature $T_\\mathrm{eff}$ [K]')
ax.set_ylabel('Luminosity $L/L_\\odot$')
ax.set_yscale('log')
ax.set_title('Stefan–Boltzmann: $L = 4\\pi R^2 \\sigma T^4$  (fixed $R = R_\\odot$)')
ax.legend()
ax.xaxis.set_minor_locator(AutoMinorLocator())

# Right panel: L vs T for the spectral types (with their MS radii)
ax2 = axes[1]
cmap_spt = plt.cm.RdYlBu_r
for i, (spt, (Teff, R_R, L_W, L_ratio)) in enumerate(luminosities.items()):
    col = cmap_spt(i / (len(luminosities) - 1))
    ax2.scatter(Teff, L_ratio, color=col, s=150, zorder=10,
                edgecolors='k', lw=0.8, label=f'{spt}  ({Teff:,} K)')
    ax2.annotate(spt, xy=(Teff, L_ratio),
                 xytext=(Teff * 1.05, L_ratio * 1.3),
                 fontsize=9, arrowprops=dict(arrowstyle='->', lw=0.6))

# Approximate ZAMS line for reference: L ~ (T/T_sun)^8 * (R/R_sun)^2 is not
# exact; instead just connect the spectral-type points as a guide.
T_spt = np.array([v[0] for v in luminosities.values()])
L_spt = np.array([v[3] for v in luminosities.values()])
idx   = np.argsort(T_spt)
ax2.plot(T_spt[idx], L_spt[idx], 'k--', lw=1, alpha=0.5, label='ZAMS (Pecaut & Mamajek 2013)')
ax2.scatter([T_sun], [1.0], color='gold', s=200, zorder=10,
            edgecolors='k', lw=0.8)
ax2.set_xlabel('Effective temperature $T_\\mathrm{eff}$ [K]')
ax2.set_ylabel('Luminosity $L/L_\\odot$')
ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.invert_xaxis()
ax2.set_title('Spectral-type luminosities (MS radii, Stefan–Boltzmann)')
ax2.legend(fontsize=8, loc='upper left')

plt.tight_layout()
plt.savefig('stefan_boltzmann_luminosity.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Synthetic photometry: computing B−V colour indices

To compute the $B-V$ colour index from a blackbody spectrum, we need to define the **filter transmission curves** $T_B(\lambda)$ and $T_V(\lambda)$, then compute the flux in each band:

$$F_B = \int_0^\infty B_\lambda(T) \, T_B(\lambda) \, d\lambda
\qquad
F_V = \int_0^\infty B_\lambda(T) \, T_V(\lambda) \, d\lambda$$

The **synthetic magnitude** in band $X$ with respect to a reference star (here Vega = A0V) is:

$$m_X = -2.5 \log_{10}\!\left(\frac{F_X}{F_{X,\rm Vega}}\right)$$

and the colour index:

$$B-V = m_B - m_V = -2.5 \log_{10}\!\left(\frac{F_B / F_{B,\rm Vega}}{F_V / F_{V,\rm Vega}}\right)$$

We model each filter as a **Gaussian transmission curve**:
- $T_B(\lambda) = \exp\left[-\tfrac{1}{2}\left(\tfrac{\lambda - 440\,\rm nm}{40\,\rm nm}\right)^2\right]$
- $T_V(\lambda) = \exp\left[-\tfrac{1}{2}\left(\tfrac{\lambda - 550\,\rm nm}{40\,\rm nm}\right)^2\right]$

```python
# ── Define Gaussian filter transmission curves ────────────────────────────
def gaussian_filter(lam_nm, center_nm, sigma_nm):
    """Gaussian bandpass filter transmission (peak=1)."""
    return np.exp(-0.5 * ((lam_nm - center_nm) / sigma_nm)**2)


# Filter parameters
B_center, B_sigma = 440.0, 40.0   # nm
V_center, V_sigma = 550.0, 40.0   # nm

# Wavelength grid for integration (finer near the filters)
lam_int_nm = np.linspace(300, 850, 10000)
lam_int_m  = lam_int_nm * 1e-9

# Transmission curves
T_B = gaussian_filter(lam_int_nm, B_center, B_sigma)
T_V = gaussian_filter(lam_int_nm, V_center, V_sigma)

# ── Visualise the filters ──────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 4))

# Shade visible spectrum
ax.axvspan(380, 700, color='lightyellow', alpha=0.7, zorder=0, label='Visible')

# Plot filters
ax.fill_between(lam_int_nm, T_B, alpha=0.4, color='royalblue', label=f'B filter (centre={B_center} nm, σ={B_sigma} nm)')
ax.fill_between(lam_int_nm, T_V, alpha=0.4, color='forestgreen', label=f'V filter (centre={V_center} nm, σ={V_sigma} nm)')
ax.plot(lam_int_nm, T_B, 'royalblue', lw=1.5)
ax.plot(lam_int_nm, T_V, 'forestgreen', lw=1.5)

# Overplot normalised Planck spectra for the Sun
B_sun = B_lambda(lam_int_m, T_sun)
B_sun_norm = B_sun / B_sun.max() * 0.9
ax.plot(lam_int_nm, B_sun_norm, 'goldenrod', lw=2, ls='--',
        label=f'Sun ($T = {T_sun:.0f}$ K, normalised)')

ax.set_xlabel('Wavelength $\\lambda$ [nm]')
ax.set_ylabel('Filter transmission (arbitrary)')
ax.set_title('Synthetic B and V photometric filters (Gaussian approximation)')
ax.set_xlim(300, 850)
ax.legend()
ax.xaxis.set_minor_locator(AutoMinorLocator())
plt.tight_layout()
plt.savefig('bv_filters.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
def compute_band_flux(T_K, filter_transmission, lam_m):
    """
    Compute synthetic flux in a photometric band by integrating B_lambda * T(lambda).

    Parameters
    ----------
    T_K : float
        Stellar temperature [K]
    filter_transmission : array
        Filter transmission curve (same grid as lam_m)
    lam_m : array
        Wavelength grid [m]

    Returns
    -------
    flux : float
        Integrated flux [W m^-2 sr^-1] (proportional units — absolute calibration not needed)
    """
    integrand = B_lambda(lam_m, T_K) * filter_transmission
    return np.trapezoid(integrand, lam_m)


def compute_BV(T_K, T_ref_Vega=9750.0):
    """
    Compute synthetic B-V colour index for a blackbody of temperature T_K.

    The colour index is calibrated so that B-V = 0 for a Vega-like A0 star
    (T_ref_Vega = 9750 K by convention from Pecaut & Mamajek 2013).

    Parameters
    ----------
    T_K : float or array
        Stellar effective temperature [K]
    T_ref_Vega : float
        Reference temperature for B-V = 0 (A0V, default 9750 K)

    Returns
    -------
    BV : float or array
        Synthetic B-V colour index [mag]
    """
    scalar = np.isscalar(T_K)
    T_arr  = np.atleast_1d(np.asarray(T_K, dtype=float))

    # Reference (Vega) fluxes
    F_B_vega = compute_band_flux(T_ref_Vega, T_B, lam_int_m)
    F_V_vega = compute_band_flux(T_ref_Vega, T_V, lam_int_m)

    BV_arr = np.empty_like(T_arr)
    for i, Ti in enumerate(T_arr):
        F_B = compute_band_flux(Ti, T_B, lam_int_m)
        F_V = compute_band_flux(Ti, T_V, lam_int_m)
        # B-V = -2.5 * log10((F_B/F_B_vega) / (F_V/F_V_vega))
        BV_arr[i] = -2.5 * np.log10((F_B / F_B_vega) / (F_V / F_V_vega))

    return BV_arr[0] if scalar else BV_arr


# ── Compute B-V for the spectral types ────────────────────────────────────
print(f'Synthetic B-V colour indices (Vega zero-point: T_A0 = 9750 K):')
print(f"\n{'Spt':>4}  {'T_eff [K]':>10}  {'B-V (Pecaut)':>14}  {'B-V (synth)':>12}  {'Diff':>8}")
print('-' * 60)

for spt, (Teff, BV_tab, R_R) in SPTYPES.items():
    BV_synth = compute_BV(Teff)
    diff = BV_synth - BV_tab
    print(f"{spt:>4}  {Teff:>10.0f}  {BV_tab:>14.3f}  {BV_synth:>12.3f}  {diff:>8.3f}")

print('\nNote: differences arise because real filter curves differ from Gaussians.')
print('The trend and sign are correct: hotter stars have smaller (bluer) B-V.')
```

## 8. The Ballesteros (2012) formula: B−V → T_eff

Ballesteros (2012, EPL 97, 34008; arXiv:1201.1809) derived an analytic expression to convert $B-V$ to $T_\mathrm{eff}$:

$$T_\mathrm{eff} = 4600 \left( \frac{1}{0.92(B-V) + 1.7} + \frac{1}{0.92(B-V) + 0.62} \right) \text{ K}$$

This is derived by assuming rectangular bandpasses and using the Wien approximation ($e^x - 1 \approx e^x$ for $x \gg 1$). Despite its simplicity it works well across the full stellar range.

The **inverse** — converting $T_\mathrm{eff}$ to $B-V$ — can be found numerically.

```python
from scipy.optimize import brentq


def ballesteros_T(BV):
    """
    Convert B-V colour index to effective temperature.
    Ballesteros (2012), EPL 97, 34008.  arXiv:1201.1809

    Parameters
    ----------
    BV : float or array
        B-V colour index [mag]

    Returns
    -------
    T : float or array
        Effective temperature [K]
    """
    BV = np.asarray(BV, dtype=float)
    return 4600.0 * (1.0 / (0.92 * BV + 1.7) + 1.0 / (0.92 * BV + 0.62))


def ballesteros_BV(T_K):
    """
    Convert effective temperature to B-V colour index (numerical inversion
    of the Ballesteros formula).

    Parameters
    ----------
    T_K : float
        Effective temperature [K]

    Returns
    -------
    BV : float
        B-V colour index [mag]
    """
    # Solve: ballesteros_T(BV) = T_K
    # B-V is monotonically decreasing with T, so root is unique in [-0.4, 2.0]
    func = lambda bv: ballesteros_T(bv) - T_K
    try:
        return brentq(func, -0.4, 2.5)
    except ValueError:
        return np.nan


# ── Grid of B-V values ─────────────────────────────────────────────────────
BV_grid = np.linspace(-0.35, 2.0, 500)
T_ball  = ballesteros_T(BV_grid)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left panel: T_eff vs B-V
ax = axes[0]
ax.plot(BV_grid, T_ball, 'navy', lw=2.5, label='Ballesteros (2012) formula')

# Mark spectral types
cmap_spt = plt.cm.RdYlBu_r
for i, (spt, (Teff, BV_tab, R_R)) in enumerate(SPTYPES.items()):
    col = cmap_spt(i / (len(SPTYPES) - 1))
    ax.scatter(BV_tab, Teff, color=col, s=120, zorder=10, edgecolors='k', lw=0.8)
    ax.annotate(spt, xy=(BV_tab, Teff),
                xytext=(BV_tab + 0.05, Teff * 1.05),
                fontsize=9, ha='left',
                arrowprops=dict(arrowstyle='->', lw=0.6))

ax.axvline(0.0, color='gray', lw=0.8, ls='--', label='B-V=0 (Vega/A0 reference)')
ax.set_xlabel('$B-V$ colour index [mag]')
ax.set_ylabel('Effective temperature $T_\\mathrm{eff}$ [K]')
ax.set_title('Ballesteros (2012) formula: $B-V$ → $T_\\mathrm{eff}$')
ax.legend()
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())

# Right panel: Round-trip accuracy
T_test = np.linspace(3000, 40000, 300)
BV_test = np.array([ballesteros_BV(T) for T in T_test])
T_roundtrip = ballesteros_T(BV_test)
residual_K  = T_roundtrip - T_test

ax2 = axes[1]
ax2.plot(T_test, residual_K, 'darkred', lw=2)
ax2.axhline(0, color='k', lw=0.8, ls='--')
ax2.set_xlabel('Input temperature $T_\\mathrm{eff}$ [K]')
ax2.set_ylabel('Round-trip residual $T_\\mathrm{out} - T_\\mathrm{in}$ [K]')
ax2.set_title('Numerical inversion accuracy: $T \\to B-V \\to T$')
ax2.xaxis.set_minor_locator(AutoMinorLocator())
ax2.yaxis.set_minor_locator(AutoMinorLocator())

plt.tight_layout()
plt.savefig('ballesteros_formula.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'Maximum round-trip error: {np.nanmax(np.abs(residual_K)):.4f} K')
```

## 9. The colour–temperature diagram: locating spectral types

We now combine everything into a comprehensive diagram showing:
1. The Ballesteros $T_\mathrm{eff}$–$(B-V)$ relation
2. The spectral-type reference points
3. The synthetic $B-V$ from Planck integration
4. The visible colour of each star as a background gradient

```python
# Compute synthetic B-V for a fine grid of temperatures
T_synth_grid = np.linspace(3000, 40000, 80)   # 80 points (quad per point)
BV_synth_grid = compute_BV(T_synth_grid)

fig, ax = plt.subplots(figsize=(11, 6))

# ── Background: approximate star colour as function of T (artistic) ─────────
# Simple RGB approximation: very hot = blue, solar = yellow, cool = red
for T_bar, bv_bar in zip(T_synth_grid, BV_synth_grid):
    frac = np.clip((T_bar - 3000) / (40000 - 3000), 0, 1)
    if frac > 0.7:       # hot, blue-white
        r, g, b = 0.5 + 0.5 * frac, 0.5 + 0.5 * frac, 1.0
    elif frac > 0.4:     # solar, yellow-white
        r, g, b = 1.0, 0.9, 0.6 * frac
    else:               # cool, orange-red
        r, g, b = 1.0, 0.3 + 0.5 * frac, 0.1
    ax.axvline(bv_bar, color=(r, g, b), alpha=0.12, lw=4)

# ── Ballesteros formula ───────────────────────────────────────────────────
ax.plot(BV_grid, T_ball / 1000, 'k-', lw=2.5, label='Ballesteros (2012)', zorder=5)

# ── Synthetic B-V from Planck integration ────────────────────────────────
mask_valid = np.isfinite(BV_synth_grid) & (BV_synth_grid > -0.5)
ax.scatter(BV_synth_grid[mask_valid], T_synth_grid[mask_valid] / 1000,
           color='steelblue', s=30, alpha=0.6, zorder=6,
           label='Synthetic (Planck + Gaussian filters)')

# ── Spectral type reference points ───────────────────────────────────────
cmap_spt = plt.cm.RdYlBu_r
for i, (spt, (Teff, BV_tab, R_R)) in enumerate(SPTYPES.items()):
    col = cmap_spt(i / (len(SPTYPES) - 1))
    ax.scatter(BV_tab, Teff / 1000, color=col, s=250, zorder=15,
               edgecolors='k', lw=1.2,
               label=f'{spt}  ($T={Teff/1000:.1f}$ kK)')
    ax.annotate(spt, xy=(BV_tab, Teff / 1000),
                xytext=(BV_tab - 0.18, Teff / 1000 + 0.5),
                fontsize=10, fontweight='bold', ha='center',
                arrowprops=dict(arrowstyle='->', lw=0.8))

# Sun marker
BV_sun = ballesteros_BV(T_sun)
ax.scatter(BV_sun, T_sun / 1000, color='gold', s=350, zorder=20,
           edgecolors='k', lw=1.5, marker='*',
           label=f'Sun ($B-V={BV_sun:.2f}$, $T={T_sun:.0f}$ K)')

ax.set_xlabel('$B-V$ colour index [mag]  (blue $\\leftarrow$ hotter | cooler $\\rightarrow$ red)', fontsize=13)
ax.set_ylabel('Effective temperature $T_\\mathrm{eff}$ [kK]', fontsize=13)
ax.set_title('Colour–temperature diagram: spectral types on the Ballesteros relation', fontsize=13)
ax.set_xlim(-0.5, 2.1)
ax.set_ylim(0, 45)
ax.legend(fontsize=8.5, ncol=2, loc='upper right')
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())

plt.tight_layout()
plt.savefig('color_temperature_diagram.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 10. Fraction of solar radiation in the visible

What fraction of the Sun's total radiation falls in the visible band (380–700 nm)?

```python
# Fine wavelength grid for integration
lam_fine_m = np.linspace(1e-9, 5e-6, 200000)   # 1 nm to 5000 nm

# Full-spectrum flux
B_sun_fine = B_lambda(lam_fine_m, T_sun)
F_total    = np.trapezoid(B_sun_fine, lam_fine_m)

# Visible band (380–700 nm)
mask_vis  = (lam_fine_m >= 380e-9) & (lam_fine_m <= 700e-9)
F_visible = np.trapezoid(B_sun_fine[mask_vis], lam_fine_m[mask_vis])

# UV (<380 nm)
mask_uv   = lam_fine_m < 380e-9
F_uv      = np.trapezoid(B_sun_fine[mask_uv], lam_fine_m[mask_uv])

# Infrared (>700 nm)
mask_ir   = lam_fine_m > 700e-9
F_ir      = np.trapezoid(B_sun_fine[mask_ir], lam_fine_m[mask_ir])

print('Solar blackbody spectrum energy fractions:')
print(f'  UV     (<380 nm): {F_uv/F_total*100:.1f}%')
print(f'  Visible (380-700 nm): {F_visible/F_total*100:.1f}%')
print(f'  Infrared (>700 nm): {F_ir/F_total*100:.1f}%')
print(f'  Sum: {(F_uv + F_visible + F_ir)/F_total*100:.3f}%')
print()
print('Physical interpretation:')
print('  The Sun emits ~44% of its energy in the visible — our eyes evolved to be')
print('  sensitive in exactly this window of peak solar emission.')

# Repeat for other temperatures
print(f'\n{"Temp [K]":>10}  {"UV%":>8}  {"Vis%":>8}  {"IR%":>8}')
print('-' * 42)
for T_test in [3000, 5778, 10000, 30000]:
    B_t = B_lambda(lam_fine_m, T_test)
    F_t = np.trapezoid(B_t, lam_fine_m)
    F_v = np.trapezoid(B_t[mask_vis], lam_fine_m[mask_vis])
    F_u = np.trapezoid(B_t[mask_uv],  lam_fine_m[mask_uv])
    F_i = np.trapezoid(B_t[mask_ir],  lam_fine_m[mask_ir])
    print(f"{T_test:>10}  {F_u/F_t*100:>8.1f}  {F_v/F_t*100:>8.1f}  {F_i/F_t*100:>8.1f}")
```

## 13. Summary

In this notebook we derived and applied the key laws governing stellar radiation:

| Topic | Key formula | Result |
|:------|:-----------:|:-------|
| Planck function | $B_\lambda = \frac{2hc^2}{\lambda^5}\frac{1}{e^{hc/\lambda k_B T}-1}$ | Full spectral energy distribution |
| Wien's law | $\lambda_\mathrm{max} T = 2.898 \times 10^{-3}$ m K | Peak shifts blue for hot stars |
| Stefan–Boltzmann | $L = 4\pi R^2 \sigma_{\rm SB} T^4$ | Luminosity from radius + temperature |
| B-V colour index | $B-V = -2.5\log_{10}(F_B/F_V) + C$ | Blue stars: small/negative $B-V$ |
| Ballesteros formula | $T = 4600(\frac{1}{0.92(B-V)+1.7}+\frac{1}{0.92(B-V)+0.62})$ | Accurate $B-V \to T$ conversion |

**Key numbers to remember:**
- Sun: $T_\odot = 5778$ K, $\lambda_\mathrm{max} = 501$ nm, $L_\odot = 3.828 \times 10^{26}$ W, $B-V \approx 0.65$
- Spectral sequence O→M: temperature decreases from ~40 000 K to ~3 500 K
- $B-V$ ranges from $\approx -0.33$ (O) to $\approx +1.6$ (M)
- The Sun emits ~44% of its energy in the visible range
